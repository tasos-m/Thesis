# Association Rule Learning - Apriori Algorithm

```python
import pandas as pd
from apyori import apriori
```

#### Load the dataset saved as ontology

```python
name = 'ao-aa-f'
dataset = pd.read_csv(f'C:/Users/anast/Desktop/Thesis/MachineLearning/Ontology/DatasetOntology/{name}.txt', header=None, sep=',', quoting=3)
```

#### Transform the input data in transactions format for the apriori algorithm

```python
# the bigger transaction should be in the first row for a proper run
transactions = []
num = 744 # changes based on the input
for i in range(0, dataset.__len__()):
    transactions.append([str(dataset.values[i, j]) for j in range(0, num)])
```

####  Training the Apriori model on the dataset


```python
rules = apriori(transactions=transactions, min_support=0.05, min_confidence=0.05, min_lift=1.01, min_length=2, max_length=2)
```

```python
results = list(rules)
```

```python
# Putting the results well organised into a Pandas DataFrame
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
resultsinDataFrame.drop(columns={'Lift'}, inplace=True)
```

```python
resultsinDataFrame.to_csv(f'C:/Users/anast/Desktop/results-{name}.csv', header=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence'], index=False)
print(resultsinDataFrame.nlargest(n = 20, columns = ['Support', 'Confidence']))
```
